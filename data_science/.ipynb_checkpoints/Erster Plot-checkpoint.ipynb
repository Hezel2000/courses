# Mein erster Plot

```python
import numpy as np
import matplotlib.pyplot as plt
```

# Scatter Plot

```python
x = [1,2,3,4]
y = [5,6,7,8]


fig, ax = plt.subplots()

plt.scatter(x, y)
plt.ylim(0,10)
plt.xlim(-3, 8)
plt.xlabel('Si (Gew%)')
plt.ylabel('Mg (Gew%)')
plt.legend('Mg (Gew%)')

ax.xaxis.set_ticks_position('both')
ax.yaxis.set_ticks_position('both')
ax.minorticks_on()
ax.tick_params(which = 'major', length = 7, width = 1, direction = 'in')
ax.tick_params(which = 'minor', length = 4, width = 1, direction = 'in')

plt.show()

fig.savefig('test.pdf')
```

# Function

```python
x = np.arange(-2, 8, .1)
y = np.sin(x)

plt.plot(x, y)

plt.show()
```
