## Practise

Show an interactive decay plot with the half-life as variable. In a plot next to this, shwo the increase of the daughter isotope.<br>
The x-axis shall go from 0 to the age of the Earth.<br>
Suppress the appearing warnings.

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact
import warnings
warnings.filterwarnings('ignore')
```

```python
def radDecay(halfLife):
    t = np.linspace(0, 4.56)  #t in Ga

    decay = np.exp(-np.log(2)/halfLife * t)
    growth = 1 - decay

    fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (9, 6))

    ax1 = plt.subplot(2, 2, 1)
    ax2 = plt.subplot(2, 2, 2)

    ax1.plot(t, decay)
    ax2.plot(t, growth)
    ax1.set_xlabel('t (Ga)')
    ax2.set_xlabel('t (Ga)')
    ax1.set_ylabel('decay')
    ax2.set_ylabel('growth')
    ax1.set_xlim(0, 4.56)
    ax2.set_xlim(0, 4.56)
    ax1.set_ylim(0, 1.1)
    ax2.set_ylim(0, 1.1)

    plt.show()

interact(radDecay, halfLife = (0, 5, .1))
```
