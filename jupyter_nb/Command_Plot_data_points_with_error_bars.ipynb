# Command– Plot data points with error bars

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
df = pd.read_excel('data/Fe isotope data.xlsx')
df.columns.tolist()
```

```python
slope, intercept, rvalue, pvalue, stderr = stats.linregress(df['δ56/54Fe'], df['δ57/54Fe'])
xReg = np.linspace(-.8, .4)
yReg = slope * xReg + intercept

xData = 'δ56/54Fe'
yData = 'δ57/54Fe'

plt.plot(xReg, yReg, '--', color = 'grey', lw = 2, zorder = 1)
plt.errorbar(df[xData], df[yData]
             ,xerr = .05 * df['δ56/54Fe']
             ,yerr = df['2 sd (57)']
             ,fmt = 'o'
             ,color = 'black'
             ,elinewidth = .6
             ,capsize = 2
             ,zorder = 2)
plt.xlabel(xData)
plt.ylabel(yData)
```
