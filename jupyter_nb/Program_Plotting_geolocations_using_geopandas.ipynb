# Plotting geolocations using geopandas

```python
import pandas as pd
import geopandas as gpd
```

```python
world_data = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
world_data.plot(color = 'yellow', edgecolor = 'grey', linewidth = .1)
```

```python
data = 'data/West African Craton.csv'
df = pd.read_csv(data)

df_geo = gpd.GeoDataFrame(df, geometry = gpd.points_from_xy(df['Longitude (Min)'],df['Latitude (Min)']))

df_geo.plot(color = 'brown', markersize = 50)
```

```python
axis = world_data.plot(color = 'yellow', edgecolor = 'grey', linewidth = .1)
df_geo.plot(ax = axis)
```

```python
world_data = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
axis = world_data.plot(color = 'yellow', edgecolor = 'grey', linewidth = .1)
df_geo.plot(ax = axis)
```
