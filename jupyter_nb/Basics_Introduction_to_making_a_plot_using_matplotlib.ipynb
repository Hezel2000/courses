# Basics- Introduction to making a plot using matplotlib

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dfMinData = pd.read_excel('data/Mineral Data.xlsx')
dfMyData = pd.read_csv('data/my epma data.csv')
```

## A simple line plot

```python
el = 'MnO'

plt.plot(dfMinData[el])

plt.show()
```

# A simple scatter plot

```python
el1 = 'MgO'
el2 = 'SiO2'

plt.scatter(dfMinData[el1], dfMinData[el2])

plt.show()
```

# Formating Elements

```python
el1 = 'MnO'
el2 = 'SiO2'

plt.scatter(dfMinData[el1], dfMinData[el2], label = 'Min Data')
plt.scatter(dfMyData[el1], dfMyData[el2], label = 'My Data')
plt.xlabel(el1)
plt.ylabel(el2)
plt.legend(loc = 'lower right')
plt.xlim(0, 0.22)
plt.ylim(0, 60)

plt.show()
```
